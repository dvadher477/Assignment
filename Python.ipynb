##For Loop

Use a for loop when you need to iterate over a collection (like a list, tuple, or dictionary) or any iterable object. For loops are straightforward and typically used when you know the number of iterations in advance.

```python
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print(num)
```

##While Loop

While loops are useful when you want to repeat an action until a specific condition becomes false.These loops are suitable when you're uncertain about the number of iterations needed beforehand, such as waiting for user input that meets certain criteria.

```python
num = 1
while num <= 5:
    print(num)
    num += 1
```

#**Q2**

##For Loop

```python
sum_for = 0
product_for = 1

for i in range(1, 11):
    sum_for += i
    product_for *= i

print("Using for loop:")
print("Sum of the first 10 natural numbers:", sum_for)
print("Product of the first 10 natural numbers:", product_for)
```

##While Loop

```python
sum_while = 0
product_while = 1
count = 1

while count <= 10:
    sum_while += count
    product_while *= count
    count += 1

print("\nUsing while loop:")
print("Sum of the first 10 natural numbers:", sum_while)
print("Product of the first 10 natural numbers:", product_while)
```

```python
def calculate_electricity_bill(units_consumed):
    total_bill = 0

    if units_consumed <= 100:
        total_bill = units_consumed * 4.5
    else:
        total_bill += 100 * 4.5
        units_consumed -= 100

        if units_consumed <= 100:
            total_bill += units_consumed * 6
        else:
            total_bill += 100 * 6
            units_consumed -= 100

            if units_consumed <= 100:
                total_bill += units_consumed * 10
            else:
                total_bill += 100 * 10
                units_consumed -= 100

                total_bill += units_consumed * 20

    return total_bill

units = int(input("Enter the units consumed: "))
bill = calculate_electricity_bill(units)
print(f"Electricity bill for {units} units: Rs. {bill:.2f}")
```

#**Q4**

##For Loop

```python
numbers_list_for = []

for i in range(1, 101):
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        numbers_list_for.append(i)

print("Using for loop:")
print(numbers_list_for)
```

##While Loop

```python
numbers_list_while = []

count = 1
while count <= 100:
    cube = count ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        numbers_list_while.append(count)
    count += 1

print("\nUsing while loop:")
print(numbers_list_while)
```

#**Q5**

```python
string = "I want to become a data scientist"

vowels = {'a', 'e', 'i', 'o', 'u'}

string_lower = string.lower()

vowel_count = 0

for char in string_lower:
    if char in vowels:
        vowel_count += 1

print("Total number of vowels in the string:", vowel_count)
```
